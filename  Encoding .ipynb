```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({'Color':['Blue','Red','Green','Blue','Green','Blue','Blue'],'Target':[1,0,1,1,1,0,1]})
df
```

```python
le = LabelEncoder()
df['Color'] = le.fit_transform(df['Color'])
df
```

```python
df = pd.DataFrame({"Color":['Blue','Red','Green','Blue','Green','Blue','Blue'],'Target':[1,0,1,1,1,0,1]})
df
```

```python
# one hot encoding
df_ohe = pd.get_dummies(df["Color"],prefix = "Color")
df_combined = pd.concat([df,df_ohe],axis=1)
df_combined 
```

```python
df_ohe = pd.get_dummies(df['Color'],prefix='Color',dtype=int)
df_combined = pd.concat ([df, df_ohe],axis=1)
df_combined 
```

```python
# Target Encoding 
mean_target = df.groupby('Color')['Target'].mean()
df["Encode_target"] = df['Color'].map(mean_target)
df
```
